```python
"""
@author: Earl Tan
"""

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('dataset.csv')
#cc_num = df['cc_num'].nunique()
#print(cc_num)
df['job'] = df['job'].apply(str.upper) #Capitalizes each Job Title

def getAvg():
    df['HOM_entertainment_avg'] = df['HOM_entertainment']/df['HOM_entertainment_freq']
    df['HOM_food_dining_avg'] = df['HOM_food_dining']/df['HOM_food_dining_freq']
    df['HOM_gas_transport_avg'] = df['HOM_gas_transport']/df['HOM_gas_transport_freq']
    df['HOM_grocery_net_avg'] = df['HOM_grocery_net']/df['HOM_grocery_net_freq']
    df['HOM_grocery_pos_avg'] = df['HOM_grocery_pos']/df['HOM_grocery_pos_freq']
    df['HOM_health_fitness_avg'] = df['HOM_health_fitness']/df['HOM_health_fitness_freq']
    df['HOM_home_avg'] = df['HOM_home']/df['HOM_home_freq']
    df['HOM_kids_pets_avg'] = df['HOM_kids_pets']/df['HOM_kids_pets_freq']
    df['HOM_misc_net_avg'] = df['HOM_misc_net']/df['HOM_misc_net_freq']
    df['HOM_misc_pos_avg'] = df['HOM_misc_pos']/df['HOM_misc_pos_freq']
    df['HOM_personal_care_avg'] = df['HOM_personal_care']/df['HOM_personal_care_freq']
    df['HOM_shopping_net_avg'] = df['HOM_shopping_net']/df['HOM_shopping_net_freq']
    df['HOM_shopping_pos_avg'] = df['HOM_shopping_pos']/df['HOM_shopping_pos_freq']
    df['HOM_travel_avg'] = df['HOM_travel']/df['HOM_travel_freq']

def cutColumns(): #Removes HOM_entertainment and freq, etc etc
    df = df.drop(columns = ['HOM_entertainment','HOM_entertainment_freq'], axis = 1)
    df = df.drop(columns = ['HOM_food_dining','HOM_food_dining_freq'], axis = 1)
    df = df.drop(columns = ['HOM_gas_transport','HOM_gas_transport_freq'], axis = 1)
    df = df.drop(columns = ['HOM_grocery_net','HOM_grocery_net_freq'], axis = 1)
    df = df.drop(columns = ['HOM_grocery_pos','HOM_grocery_pos_freq'], axis = 1)
    df = df.drop(columns = ['HOM_health_fitness','HOM_health_fitness_freq'], axis = 1)
    df = df.drop(columns = ['HOM_home','HOM_home_freq'], axis = 1)
    df = df.drop(columns = ['HOM_kids_pets','HOM_kids_pets_freq'], axis = 1)
    df = df.drop(columns = ['HOM_misc_net','HOM_misc_net_freq'], axis = 1)
    df = df.drop(columns = ['HOM_misc_pos','HOM_misc_pos_freq'], axis = 1)
    df = df.drop(columns = ['HOM_personal_care','HOM_personal_care_freq'], axis = 1)
    df = df.drop(columns = ['HOM_shopping_net','HOM_shopping_net_freq'], axis = 1)
    df = df.drop(columns = ['HOM_shopping_pos','HOM_shopping_pos_freq'], axis = 1)
    df = df.drop(columns = ['HOM_travel','HOM_travel_freq'], axis = 1)
    
    


df['date'] = pd.to_datetime(df.dob) #Converts date of birth to date and adds a new column
df['year'] = pd.DatetimeIndex(df.date).year #Creates new columns for year and month
df['month'] = pd.DatetimeIndex(df.date).month
df['age'] = 2020 - df['year'] #Subtracts the latest year, 2020, from year of birth

conditions = [ #Creates a condition for age groups, per 10 years
    (df['age'] >= 15) & (df['age'] < 25),
    (df['age'] >= 25) & (df['age'] < 35),
    (df['age'] >= 35) & (df['age'] < 45),
    (df['age'] >= 45) & (df['age'] < 55),
    (df['age'] >= 55) & (df['age'] < 65),
    (df['age'] >= 65) & (df['age'] < 75),
    (df['age'] >= 75) & (df['age'] < 85),
    (df['age'] >= 85)
    ]
values = ['15-24', '25-34', '35-44', '45-54', '55-64',
          '65-74', '75-84', '85 and above']
df['age_bracket'] = np.select(conditions, values) #Creates an age bracket column
#That identifies the customer's age bracket

#Different categories to be used
categ_total = ['HOM_entertainment', 'HOM_food_dining', 'HOM_gas_transport', 'HOM_grocery_net', 'HOM_grocery_pos', 'HOM_health_fitness', 'HOM_home', 'HOM_kids_pets', 'HOM_misc_net', 'HOM_misc_pos', 'HOM_personal_care', 'HOM_shopping_net', 'HOM_shopping_pos', 'HOM_travel']
categ_freq = ['HOM_entertainment_freq', 'HOM_food_dining_freq', 'HOM_gas_transport_freq', 'HOM_grocery_net_freq', 'HOM_grocery_pos_freq', 'HOM_health_fitness_freq', 'HOM_home_freq', 'HOM_kids_pets_freq', 'HOM_misc_net_freq', 'HOM_misc_pos_freq', 'HOM_personal_care_freq', 'HOM_shopping_net_freq', 'HOM_shopping_pos_freq', 'HOM_travel_freq']
job_group_list = ['ENGINEERING AND ARCHITECTURE', 'SALES', 'EDUCATION', 'COMPUTING', 'BUSINESS AND MANAGEMENT', 'SCIENCE', 'MEDIC', 'ART AND LITERATURE', 'LAW', 'GOVERNMENT', 'AGRICULTURE', 'CONSTRUCTION', 'MEDIA']
categ = categ_total + categ_freq
base_df = df.groupby(['cc_num', 'job', 'gender', 'age_bracket'])[categ].sum()
#Creates another dataframe that uses credit card num, job, gender, age, along with
#Half of Month expenses for each category.
#This summarizes a single customer's information
base_df = base_df.reset_index()


age_df = base_df.groupby('age_bracket')[categ].mean() 
#Creates a separate dataframe for age brackets and categories
age_df = age_df.reset_index()

gender_df = base_df.groupby('gender')[categ].mean()
#Creates a separate dataframe for genders and categories
gender_df = gender_df.reset_index()

def condition2(row): #Function to group jobs into job groups
    match_engineering = ['ENGINEERING', 'ARCHITECT']
    match_sales = ['SALES', 'MARKET', 'DEALER', 'RETAIL', 'INSURANCE', 'TRADE', 'CLERK', 'BARRIST', 'TRADING', 'BUYER', 'BOOKSELLER']
    match_educ = ['TEACH', 'EDUCAT', 'PROFESSOR', 'ACADEMIC', 'TUTOR', 'LEARN', 'COUNSELLOR', 'LIBRAR', 'LEXI']    
    match_computing = ['PROGRAMMING', 'COMPUTER', 'IT ', 'INFORMATION', 'REPAIR', 'HARDWARE', 'SOFTWARE', 'DATABASE', 'DEVELOPER',
                       'NETWORK', 'MECHANIC']
    match_bma = ['ACCOUNTANT', 'ACCOUNTING', 'ADMIN', 'FINANC', 'MANAGE', 'HUMAN', 'ECONOM', 'BUSINE', 'INVEST', 'EXECUTIVE', 'STAFF', 'CHIEF',
                     'BOSS', 'COMPANY', 'SECRETARY', 'ANALYST', 'CONSUL', 'COMPTROLLER']
    match_science = ['SCIENT', 'RESEARCH', 'CHEM', 'ARCHAE', 'PHYSICIST', 'BIOLO', 'SEISM', 'METALLUR', 'GEOLO', 'LABORA', 'OCEANO']
    match_medic = ['DOCTOR', 'THERAP', 'NURSE', 'PHYSICIAN', 'PSYCHO', 'EMERG', 'HEALTH', 'NUTRIT', 'HOSPIT', 'AMBU', 'PHARMA', 'MEDIC', 'PHYSIO',
                       'DENTAL', 'DENTIST', 'SURGEON', 'VETERI', 'IMMUNO', 'OPTIC', 'AID', 'ONCOL', 'OPTO', 'ORTHO', 'OSTEO', 'PATHO', 'RADIOG']
    match_artlit = ['ART', 'DESIGN', 'WRITE', 'MUSIC', 'JOURN', 'MUSEUM', 'VIDEO', 'PHOTO', 'CLOTH', 'PRODUCT', 'EDIT', 'ILLUS', 'CURATOR']
    match_law = ['LAW', 'LEGAL', 'LOBBY', 'ATTORNEY', 'PROSECUT']
    match_govern = ['GOVERN', 'POLITIC', 'FIREFIGHT', 'POLICE', 'COP', 'PRISON', 'WARDEN', 'OFFICER', 'TAX', 'AFFAIRS', 'STATISTIC', 'COMMUNITY',
                    'OCCUPATION', 'TOWN']
    match_agri = ['FARM', 'FISH', 'FOREST', 'AGRI', 'ANIMAL', 'BOTANY', 'HERB', 'GENETIC']
    match_construct = ['CONSTRUCT', 'SURVEY', 'FREIGHT', 'CONTRACTOR']
    match_media = ['BROADCAST', 'PRODUCER', 'CAMERA', 'TELEVISI', 'MEDIA', 'PRES']
    if any([x in row['job'] for x in match_engineering]):
        return 'ENGINEERING AND ARCHITECTURE'
    elif any([x in row['job'] for x in match_sales]):
        return 'SALES'
    elif any([x in row['job'] for x in match_educ]):
        return 'EDUCATION'
    elif any([x in row['job'] for x in match_computing]):
        return 'COMPUTING'
    elif any([x in row['job'] for x in match_bma]):
        return 'BUSINESS AND MANAGEMENT'
    elif any([x in row['job'] for x in match_science]):
        return 'SCIENCE'
    elif any([x in row['job'] for x in match_medic]):
        return 'MEDICINE'
    elif any([x in row['job'] for x in match_artlit]):
        return 'ART AND LITERATURE'
    elif any([x in row['job'] for x in match_law]):
        return 'LAW'
    elif any([x in row['job'] for x in match_govern]):
        return 'GOVERNMENT'
    elif any([x in row['job'] for x in match_agri]):
        return 'AGRICULTURE'
    elif any([x in row['job'] for x in match_construct]):
        return 'CONSTRUCTION'
    elif any([x in row['job'] for x in match_media]):
        return 'MEDIA'
    else:
        return 'OTHERS'

base_df['job_group'] = base_df.apply(condition2, axis = 1) #Creates a job group column
#That identifies the customer's job group

job_df = base_df.groupby('job_group')[categ].mean()
#Creates a separate dataframe for job groups and categories
job_df = job_df.reset_index()

#Line Plot
categ_plot = categ_freq # change this to categ_total
ax = age_df.plot(x = 'age_bracket', y = categ_plot, kind = "line", figsize = (15, 10))
plt.xlabel("Age Bracket")
plt.ylabel("Average spent on each category")
plt.title("Average spent per age bracket")
plt.show()
#Bar Plot
ax = gender_df.plot(x ='gender', y=categ_plot, kind = "bar", figsize = (10, 5), stacked = True)
plt.xlabel("Gender")
plt.ylabel("Average spent on each category")
plt.title("Average spent per gender")
plt.show()
#Bar Plot
for j in job_group_list:
    ax = job_df[job_df['job_group'] == 'CONSTRUCTION'].plot(x = 'job_group', y=categ_plot, kind="bar", figsize = (15, 10), stacked = False)

#Have a method that changes this to other names
plt.xlabel("Job Group")
plt.ylabel("Average spent on each category")
plt.title("Average spent per job group")
plt.show()

#Prints minimum spent per category
for c in categ_total:
    print(c + ': ' + job_df.iloc[job_df[c].idxmin()]['job_group'])
    print('With an average spending of: ' + str(job_df.iloc[job_df[c].idxmin()][c]))
    print('\n')
#Prints maximum per category
for c in categ_total:
    print(c + ': ' + job_df.iloc[job_df[c].idxmax()]['job_group'])
    print('With an average spending of: ' + str(job_df.iloc[job_df[c].idxmax()][c]))
    print('\n')



#base_df['job_group'].value_counts()
```
